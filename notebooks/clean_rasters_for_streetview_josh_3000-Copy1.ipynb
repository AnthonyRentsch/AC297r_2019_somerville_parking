```python
from src.raster import Raster
import pandas as pd
import geopandas as gpd
import os

from os import listdir
from os.path import isfile, join
```

```python
data_dir = '../data/raster_all/'
```

```python
images = []
for file in os.listdir(data_dir):
    if file.endswith(".TIF"):
        images.append(file)
```

```python
parcels = gpd.read_file('../data/raster_all_parcels/')
address_MBL = pd.read_csv('../data/address_MBL.csv', index_col=0)
parcels_address = parcels.merge(address_MBL)
```

```python
parcels_address
```

```python
def get_name(parcel_name):
    parcel_num = int(parcel_name.replace('Somerville','').replace('.TIF',''))
    row = parcels.iloc[parcel_num]
    if row['AddNum'] is None:
        row['AddNum'] = 'unknown'
    if row['Street'] is None:
        row['Street'] = 'unknown'
    address = row['AddNum'] + ' ' +row['Street']
    address_cln = address.lower().replace('  ',' ').replace(' ','_')
    filename = address_cln + '_aerial.png'
    filename = filename.replace('/', '_')
    print(filename)
    return(filename)
```

```python
rasters_already_downloaded = [f.lower() for f in listdir("../data/aerial_all/") 
                           if isfile(join("../data/aerial_all/", f))]
```

```python
rasters_already_downloaded
```

```python
for img in images:
    r = Raster(data_dir, img)
    img_filename = get_name(img)
    if img_filename not in rasters_already_downloaded:
        r.savefig('../data/aerial_all/' + img_filename, dpi = 300)
```
