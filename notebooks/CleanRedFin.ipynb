```python
import pandas as pd
import numpy as np
pd.options.display.max_rows = 999
```

```python
df = pd.read_csv("../data/redfin_data.csv")
```

```python
df.head()
```

```python
df_wide = df.pivot_table(index = 'mbl', columns='field_name', values = 'value', aggfunc=lambda x: x)
df_wide.has_data = df_wide.has_data.apply(lambda x: True if x == 'True' else False)
```

```python
df_wide = df_wide[df_wide.has_data]
```

```python
def has_driveway(row):
    x = ""
    if type(row['Parking']) == str:
        x = x + row['Parking']
    if type(row['Parking Features']) == str:
        x = x + " " + row['Parking']
    if x == "":
        return 2
    if 'Off-Street' in x:
        return 1
    elif 'Tandem' in x:
        return 1
    elif 'Driveway' in x:
        return 1
    elif 'On-Street' in x:
        return 0
    elif 'Other' in x:
        return 0
    else:
        return 1
```

```python
df_wide['has_driveway'] = df_wide.apply(has_driveway, axis = 1)
df_wide.head()
```

```python
df_wide.groupby('has_driveway')['Parking'].apply(pd.Series.value_counts).reset_index()
```

```python
def has_garage(row):
    x = ""
    if (type(row['Garage']) == str) or (type(row['Garage / Carport Type']) == str):
        return 1
    else:
        return 0
```

```python
df_wide['has_garage'] = df_wide.apply(has_garage, axis = 1)
df_wide.head()
```

```python
df_wide['has_parking'] = ((df_wide.has_garage != 0) | (df_wide.has_driveway != 0)).astype(int)
```

```python
df_wide = df_wide.reset_index()
```

```python
df_wide['MBL']  = df_wide.mbl
df_wide = df_wide.drop('mbl', axis = 1)
```

```python
df_wide[df_wide.has_parking == 0]
```

```python
address_MBL = pd.read_csv('../data/address_MBL.csv', index_col= 0)
```

```python
df_clean = address_MBL.merge(df_wide)
```

```python
df_clean.to_csv('../data/redfin_clean.csv')
```
